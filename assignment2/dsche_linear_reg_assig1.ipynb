```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
rng = np.random.RandomState()
x = 10 * rng.rand(100)
y = 2 * x - 2 + rng.randn(100)
plt.scatter(x, y);
```

```python
rng = np.random.RandomState()
x_q = 10 * rng.rand(50)
y_q = x_q**2 - 4*x_q + 4 + rng.randn(100)
plt.scatter(x_q, y_q);
```

```python
def loss(y_hat, y):
    l = np.abs((y_hat - y)**3/3) #cancels 3 out while taking pd
    return l.mean()
def loss_2(y_hat, y):
    m = np.abs((y_hat - y)) 
    return m.mean()
def loss_3(y_hat, y):
    n = (y_hat - y)**4/4
    return n.mean()
def loss_4(y_hat, y):
    o = np.abs((y_hat - y)**7)/7 
    return o.mean()            
```

```python
L1=[]
L2=[]
L3=[]
L4=[]
N=x.shape[0]
```

```python
def forward_pass(x,m,c):
    return m*x+c
def forward_pass_q(x,a,b,d):
    return a*x**2 + b*x + d   
```

```python
m = np.random.randn(1)
a = np.random.randn(1)
b = np.random.randn(1)
```

```python
# Random Initialisation
m1 = np.random.randn(1)
c1 = np.random.randn(1)

# Hyperparameters
learningrate = 0.75*1e-6
numiter = 50000

for i in range(numiter):
    # forward pass
    y_hat=forward_pass(x,m1,c1)
    
    # Gradient
    dm = (1.0/N) *  -np.sum(np.multiply(x,(y - (m1*x +c1))**2))
    dc = (1.0/N) * -np.sum((y - (m1*x +c1))**2)
    
    m1 = m1 -learningrate * dm
    c1 = c1 - learningrate * dc
                           
    if i%5000==0:
        L1.append(loss(y_hat,y))
        print ("no of iterations:",i )
        plt.plot(x,y,'go')
        plt.plot(x,y_hat,'r*')
    #plt.plot(x,m1*x+c1,'r*')
        plt.ylabel('Data')
        plt.show()
```

```python
# Random Initialisation
m2 = np.random.randn(1)
c2 = np.random.randn(1)

# Hyperparameters
learningrate = 0.00001275
numiter = 50000

for i in range(numiter):
    # forward pass
    y_hat_2=forward_pass(x,m2,c2)
    
    # Gradient
    dm = (1.0/N) *  -np.sum(x)
    dc = (1.0/N) * -np.sum(1)
    
    m2 = m2 -learningrate * dm
    c2 = c2 - learningrate * dc
                           
    if i%5000==0:
        L2.append(loss_2(y_hat_2,y))
        print ("no of iterations:",i )
        plt.plot(x,y,'go')
        plt.plot(x,y_hat_2,'r*')
    #plt.plot(x,m1*x+c1,'r*')
        plt.ylabel('Data')
        plt.show()
```

```python
# Random Initialisation
a1 = np.random.randn(1)
b1 = np.random.randn(1)
c1 = np.random.randn(1)

# Hyperparameters
learningrate = 0.0000000007
numiter = 50000

for i in range(numiter):
    # forward pass
    y_hat_3=forward_pass_q(x_q,a1,b1,c1)
    # Loss
   
    # Gradient
    da = (1.0/N) *  -np.sum(np.multiply(x_q**2,(y_q - (a1*x_q**2 + b1*x_q +c1))**3))
    db = (1.0/N) *  -np.sum(np.multiply(x_q,(y_q - (a1*x_q**2 + b1*x_q +c1))**3))
    dc = (1.0/N) *  -np.sum((y_q - (a1*x_q**2 + b1*x_q +c1))**3)
    
    a1 = a1 - learningrate * da
    b1 = b1 - learningrate * db
    c1 = c1 - learningrate * dc

    if i%5000==0:
        L3.append(loss_3(y_hat_3,y_q))
        print ("no of iterations:",i )
        plt.plot(x_q,y_q,'go')
        plt.plot(x_q,y_hat_3,'r*')
    #plt.plot(x,m1*x+c1,'r*')
        plt.ylabel('Data')
        plt.show()
```
