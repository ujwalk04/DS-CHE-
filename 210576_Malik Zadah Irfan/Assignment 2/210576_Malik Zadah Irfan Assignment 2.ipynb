```python
#loss function |x - x|
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
rng = np.random.RandomState(10)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
# Scale the data
x_mean = np.mean(x)
x_std = np.std(x)
y_mean = np.mean(y)
y_std = np.std(y)
x = (x - x_mean) / x_std
y = (y - y_mean) / y_std

# Define the loss function
def loss(y_hat, y):
    return np.sum(np.abs(y_hat - y) ** 3) / y.shape[0]

# Define the forward pass function
def forward_pass(x, m, c):
    return m * x + c

# Set the hyperparameters
learning_rate = 0.01
n_iterations = 50000

# Initialize the model parameters
m = np.random.randn(1)
c = np.random.randn(1)

# Fit the model
losses = []
for i in range(n_iterations):
    # Make predictions
    y_hat = forward_pass(x, m, c)
    
    # Compute the loss
    l = loss(y_hat, y)
    losses.append(l)
    
    # Print the loss and plot the predictions every 5000 iterations
    if i % 5000 == 0:
        print(f"Iteration {i}: Loss = {l}")
        plt.plot(x, y_hat, '--')
        plt.scatter(x, y)
        plt.show()
    
    # Compute the gradients
    dm = 3 * np.sum(np.abs(y_hat - y) ** 2 * np.sign(y_hat - y) * x) / y.shape[0]
    dc = 3 * np.sum(np.abs(y_hat - y) ** 2 * np.sign(y_hat - y)) / y.shape[0]
    
    # Update the model parameters
    m -= learning_rate * dm
    c -= learning_rate * dc
```

```python
plt.plot(losses)
```

```python
#quadratic regression model with loss function |x - x|7
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
rng = np.random.RandomState(10)
x = 10 * rng.rand(50)
y = x ** 2 - 5 * x + rng.randn(50)

# Calculate the minimum and maximum values
x_min = np.min(x)
x_max = np.max(x)
y_min = np.min(y)
y_max = np.max(y)

# Scale the data to the range [0, 1]
x_scaled = (x - x_min) / (x_max - x_min)
y_scaled = (y - y_min) / (y_max - y_min)

# Define the loss function
def loss(y_hat, y):
    return np.sum(np.abs(y_hat - y) ** 7) / y.shape[0]

# Define the forward pass function
def forward_pass(x, a, b, c):
    return a * x ** 2 + b * x + c

# Set the hyperparameters
learning_rate = 0.01
n_iterations = 100000

# Initialize the model parameters
a = np.random.randn(1)
b = np.random.randn(1)
c = np.random.randn(1)
# Fit the model
losses = []
for i in range(n_iterations):
    # Make predictions
    y_hat = forward_pass(x_scaled, a, b, c)
    
    # Compute the loss
    l = loss(y_hat, y_scaled)
    losses.append(l)
    
    # Compute the gradients
    da = 7 * np.sum(np.abs(y_hat - y_scaled) ** 6 * x_scaled ** 2 * np.sign(y_hat - y_scaled)) / y_scaled.shape[0]
    db = 7 * np.sum(np.abs(y_hat - y_scaled) ** 6 * x_scaled *np.sign(y_hat - y_scaled)) / y_scaled.shape[0]
    dc = 7 * np.sum(np.abs(y_hat - y_scaled) ** 6*np.sign(y_hat - y_scaled)) / y_scaled.shape[0]  
    # Update the model parameters
    a -= learning_rate * da
    b -= learning_rate * db
    c -= learning_rate * dc

# Print the loss and plot the predictions every 5000 iterations
    if i % 5000 == 0:
       print(f"Iteration {i}: Loss = {l}")
       
       # Plot the data
       plt.scatter(x_scaled, y_scaled, color='blue')
       
       # Plot the prediction curve
       x_plot = np.linspace(np.min(x_scaled), np.max(x_scaled), 100)
       y_plot = forward_pass(x_plot, a, b, c)
       plt.plot(x_plot, y_plot, color='red')
       plt.show()
```

```python
# Plot the loss curve
plt.plot(losses)
plt.show()
```
