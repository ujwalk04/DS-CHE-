```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
rng = np.random.RandomState(10)
x= 10* rng.rand(50)
y = 2*x +8 +rng.rand(50)
```

```python
# Data Standardization
x_mean = np.mean(x)
x_std = np.std(x)
y_mean = np.mean(y)
y_std = np.std(y)
x = (x - x_mean) / x_std
y = (y - y_mean) / y_std

# Define the loss function |x - x|
def loss(y_hat, y):
    return np.sum(np.abs(y_hat - y) ** 3) / y.shape[0]

# Define the forward pass function
def forward_pass(x, m, c):
    return m * x + c
# Set the hyperparameters
learning_rate = 0.001
n_iterations = 20000

# Initialize the model parameters
m = np.random.randn(1)
c = np.random.randn(1)

# Fit the model
losses = []
for i in range(n_iterations):
    # Make predictions
    y_hat = forward_pass(x, m, c)
    
    # Compute the loss
    l = loss(y_hat, y)
    losses.append(l)
    
    # Print the loss and plot the predictions every 2000 iterations
    if i % 1000 == 0:
        print(f"Iteration {i}: Loss = {l}")
        plt.plot(x, y_hat, '--')
        plt.scatter(x, y)
        plt.show()
    
    # Compute the gradients
    dm = 3 * np.sum(np.abs(y_hat - y) ** 2 * np.sign(y_hat - y) * x) / y.shape[0]
    dc = 3 * np.sum(np.abs(y_hat - y) ** 2 * np.sign(y_hat - y)) / y.shape[0]
    
    # Update the model parameters
    m -= learning_rate * dm
    c -= learning_rate * dc
```

```python
plt.plot(losses)
plt.xlabel("Iteration")
plt.ylabel("Loss")
```
