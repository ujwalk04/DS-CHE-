```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("superheated_vapor_properties.csv")
df
```

```python
V_data=df[df['Property']=='V']
V_data
```

```python
dataset1=V_data[(V_data['Pressure']<300)]
dataset1
```

```python
dataset2 = V_data[(V_data['Pressure']>300) & (V_data['Pressure']<1500)]
dataset2
```

```python
dataset3=V_data[(V_data['Pressure']>1500)]
dataset3
```

```python
d1 = dataset1[['Pressure','Liq_Sat']]
d1
```

```python
d2 = dataset2[['Pressure','Liq_Sat']]
d2
```

```python
d3 = dataset3[['Pressure','Liq_Sat']]
d3
```

```python
X = d1['Pressure'].values
Y = d1['Liq_Sat'].values

x_mean = np.mean(X)
y_mean = np.mean(Y)

n = len(X)

num = 0
deno = 0
for i in range(n):
    num += (X[i] - x_mean) * (Y[i] - y_mean)
    deno += (X[i] - x_mean) ** 2
    
m = num / deno
c = y_mean - (m * x_mean)
#printing the coefficient
print(m, c)
```

```python
 
x_max = np.max(X) + 100
x_min = np.min(X) - 100

x = np.linspace(x_min, x_max, 1000)
y = c + m * x

plt.plot(x, y, color='b', label='Linear Regression',)

plt.scatter(X, Y, color='g', label='Data Point')

plt.xlabel('Pressure')

plt.ylabel('Liq_Sat')
plt.legend()
plt.show()
```

```python
r2 = 0
for i in range(n):
    y_pred=  c + m* X[i]
    r2 += (Y[i] - y_pred) ** 2
    
r2 = np.sqrt(r2/n)
print(r2)
```

```python
sumofsquares = 0
sumofresiduals = 0
for i in range(n) :
    y_pred = c + m * X[i]
    sumofsquares += (Y[i] - y_mean) ** 2
    sumofresiduals += (Y[i] - y_pred) **2
    
score  = 1 - (sumofresiduals/sumofsquares)
print(score)
```

```python
X = np.array(d1['Pressure']).reshape(-1, 1)
y = np.array(d1['Liq_Sat']).reshape(-1, 1)
  

d1.dropna(inplace = True)
  
# Dropping any rows with Nan values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
  
# Splitting the data into training and testing data
regr = LinearRegression()
  
regr.fit(X_train, y_train)
print(regr.score(X_test, y_test))
```

```python
y_pred = regr.predict(X_test)
plt.scatter(X_test, y_test, color ='b')
plt.plot(X_test, y_pred, color ='k')
  
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
  
mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)
#squared True returns MSE value, False returns RMSE value.
mse = mean_squared_error(y_true=y_test,y_pred=y_pred) #default=True
rmse = mean_squared_error(y_true=y_test,y_pred=y_pred,squared=False)
  
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```
